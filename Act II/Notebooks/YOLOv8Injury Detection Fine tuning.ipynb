```python
from ultralytics import YOLO
import os
import time
```

```python
model = YOLO('yolov8m.pt')  
```

# fine tune with best weights 

```python
yaml_path = "/kaggle/input/injured-people-in-accident/data.yaml"  
project_dir = "/kaggle/working/yolo_project"

best_weights_path = "/kaggle/input/model_weights/pytorch/default/1/best (1).pt" 
if os.path.exists(best_weights_path):
    print(f"Loading weights from {best_weights_path}...")
    model = YOLO(best_weights_path)
else:
    print("No best.pt file found. Ensure weights were saved correctly.")

print("Starting second stage of training (50 epochs)...")
train_results_stage2 = model.train(
    data=yaml_path,
    epochs=50, 
    imgsz=640,
    batch=32,
    save_period=5,
    project=project_dir,
    name="finetune_run_continued",
)
```

```python
metrics = model.val()  # Evaluate on the validation set
print(metrics.box.map)
```
