```python
# Define paths for YOLOv8 dataset
train_images_folder = '/content/drive/MyDrive/GP/train/images'
train_labels_folder = '/content/drive/MyDrive/GP/train/labels'
val_images_folder = '/content/drive/MyDrive/GP/Val/images'
val_labels_folder = '/content/drive/MyDrive/GP/Val/labels (1)'

# Temporary folders for processing into class-based folders
train_processed_folder = 'train_processed'
val_processed_folder = 'val_processed'

# Define the class index mapping
class_mapping = {
    0: 'High',
    1: 'High',
    2: 'High',
    3: 'Low'
}

# Function to classify based on the label
def classify_label(class_id):
    return 'Low' if class_id in [3] else 'High'

def process_dataset(images_folder, labels_folder, output_folder):
    # Clear and recreate the output folder
    if os.path.exists(output_folder):
        shutil.rmtree(output_folder)
    os.makedirs(os.path.join(output_folder, 'Low'))
    os.makedirs(os.path.join(output_folder, 'High'))

    # Iterate through label files
    for label_file in os.listdir(labels_folder):
        label_path = os.path.join(labels_folder, label_file)
        print(f"Processing label file: {label_path}")  # Debugging line

        # Read the label file
        with open(label_path, 'r') as f:
            lines = f.readlines()

        # Extract class indices
        class_indices = [int(line.split()[0]) for line in lines]
        print(f"Class indices found: {class_indices}")  # Debugging line

        # Determine the new label
        new_label = 'High'  # Default label
        for class_id in class_indices:
            new_label = classify_label(class_id)
            if new_label == 'Low':  # Prioritize "High" if found
                break
        print(f"Assigned label: {new_label}")  # Debugging line

        # Get the corresponding image path
        image_name = os.path.splitext(label_file)[0] + '.jpg'  # Assuming .jpg images
        image_path = os.path.join(images_folder, image_name)

        # Check if the image exists and copy it to the appropriate folder
        if os.path.exists(image_path):
            print(f"Copying image: {image_path} to {output_folder}/{new_label}")  # Debugging line
            shutil.copy(image_path, os.path.join(output_folder, new_label))
        else:
            print(f"Image not found: {image_path}")  # Debugging line

# Process the training and validation datasets
process_dataset(train_images_folder, train_labels_folder, train_processed_folder)
process_dataset(val_images_folder, val_labels_folder, val_processed_folder)

# Function to swap folder names
def swap_folders(base_path, folder1, folder2):
    folder1_path = os.path.join(base_path, folder1)
    folder2_path = os.path.join(base_path, folder2)

    temp_folder1_path = os.path.join(base_path, folder1 + "_temp")
    temp_folder2_path = os.path.join(base_path, folder2 + "_temp")

    # Swap the folder names using temporary names
    if os.path.exists(folder1_path):
        os.rename(folder1_path, temp_folder1_path)
    if os.path.exists(folder2_path):
        os.rename(folder2_path, temp_folder2_path)

    if os.path.exists(temp_folder1_path):
        os.rename(temp_folder1_path, folder2_path)
    if os.path.exists(temp_folder2_path):
        os.rename(temp_folder2_path, folder1_path)

    print(f"Swapped {folder1} and {folder2} in {base_path}")

# Swap folders in train_processed
swap_folders(train_processed_folder, 'High', 'Low')

# Swap folders in val_processed
swap_folders(val_processed_folder, 'High', 'Low')

# Define paths for the processed datasets
trainpath = train_processed_folder
valpath = val_processed_folder
print('Train processed folder:', train_processed_folder)
print('Validation processed folder:', val_processed_folder)

# Augmentation
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1.0 / 255)

train_data = train_datagen.flow_from_directory(
    trainpath,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

val_data = val_datagen.flow_from_directory(
    valpath,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary'
)

```

```python

# Fine-Tuning
backbone = EfficientNetB0(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
backbone.trainable = False

x = backbone.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dense(1, activation='sigmoid')(x)  #0 or 1
model = Model(inputs=backbone.input, outputs=x)


model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=30,
    verbose=1
)
# model.save('severity_classification.h5')
#loss, accuracy = model.evaluate(val_data)


```

```python
# #################################################### Prediction ####################################################
def predict_severity(image_path, model):
    img = image.load_img(image_path, target_size=(224, 224))
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    predictions = model.predict(img_array)
    print(print(f"Confidence value: {predictions}"))
    confidence = predictions[0][0]
    severity = 'Low' if confidence < 0.5 else 'High'
    return severity, confidence
```

```python
for layer in backbone.layers:
    layer.trainable = True

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss='binary_crossentropy', metrics=['accuracy'])

history_finetune = model.fit(
    train_data,
    validation_data=val_data,
    epochs=30,
    verbose=1
)
```
